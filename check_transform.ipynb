```python
import matplotlib.pyplot as plt
import cv2

# Krok 1: Wczytaj obraz
image_path = 'needle_augmented/train/images/1.jpg'
image = cv2.imread(image_path)  # Wczytaj obraz za pomocÄ… OpenCV
```

```python
kpts = []
ann_path = 'needle_augmented/train/labels/1.txt'

with open(ann_path, 'r') as f:
    t = f.read()
    print([float(e) for e in t.split(' ')])
    kpt = [float(e) for e in t.split(' ')]
    kpts = [kpt[5], kpt[6], kpt[8], kpt[9]]
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.scatter(kpts[0], kpts[1], c='r')
plt.scatter(kpts[2], kpts[3], c='g')
plt.show()
```
